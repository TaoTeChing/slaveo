## 设置开仓条件

### 均线开仓

```python
yestorday = datetime.date.today() - datetime.timedelta(days=1)
yes_str = yestorday.strftime("%Y-%m-%d")
ts.get_hists(codes, start=yes_str, end=yes_str)
```

```python
his = ts.get_hists(codes, start=yes_str, end=yes_str);his
```

```python
his = his[["code", "ma5", "close"]].set_index("code")
his.columns = ["ma", "close"]
his
```

```python
s.codes[["open_ma"]] = his[["ma"]]
s.codes.open_ma = his.ma
s.codes.close = his.close
s.codes
```
